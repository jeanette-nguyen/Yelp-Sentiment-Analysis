```python
import pickle

with open('rawstars.p', 'rb') as s:
    load_raw_stars = pickle.load(s)
with open("rawreviews.p", "rb") as t:
    load_raw_reviews = pickle.load(t)
```

```python
with open('first300000_star.p', 'rb') as s:
    load_stars = pickle.load(s)
with open("first300000_text.p", "rb") as t:
    load_reviews = pickle.load(t)
```

```python
review_len = [len(review[0]) for review in load_raw_reviews]
star_len = [0]*5
num_reviews = [0]*5
for i in range(len(review_len)):
    star_len[load_raw_stars[i]-1] = star_len[load_raw_stars[i]-1] + review_len[i]
    num_reviews[load_raw_stars[i]-1] = num_reviews[load_raw_stars[i]-1] + 1
for i in range(len(star_len)):
    star_len[i] = star_len[i]/num_reviews[i]
```

```python
review_len = [len(review[0].split()) for review in load_raw_reviews]
star_len = [0]*5
num_reviews = [0]*5
for i in range(len(review_len)):
    star_len[load_raw_stars[i]-1] = star_len[load_raw_stars[i]-1] + review_len[i]
    num_reviews[load_raw_stars[i]-1] = num_reviews[load_raw_stars[i]-1] + 1
for i in range(len(star_len)):
    star_len[i] = star_len[i]/num_reviews[i]
```

```python
from bokeh.plotting import figure, output_file, show

# prepare some data
x = range(1,6)
y = star_len


# create a new plot with a title and axis labels
p = figure(title="Average Length of Review Per Star", # title of figure
           x_axis_label='Star Rating', 
           y_axis_label='Length of Review in Words',
           width= 500,  # figure width
           height = 500) # figure height

# add a line renderer with legend and line thickness
p.line(x, y, 
        #text that appears in legend
       line_width=2)  # width of line

# show the results
show(p)
```

```python
word_freq = {1:{}, 2:{}, 3:{}, 4:{}, 5:{}}
for index in range(len(load_raw_reviews)):
    review = load_raw_reviews[index][0]
    review = review.split()
    for word in review:
        if word in word_freq[load_raw_stars[index]]:
            word_freq[load_raw_stars[index]][word] += 1
        else:
            word_freq[load_raw_stars[index]][word] = 1
```

```python
word_freq_processed = {1:{}, 2:{}, 3:{}, 4:{}, 5:{}}
for index in range(len(load_reviews)):
    review = load_reviews[index]
    review = review.split()
    for word in review:
        if word in word_freq_processed[load_stars[index]]:
            word_freq_processed[load_stars[index]][word] += 1
        else:
            word_freq_processed[load_stars[index]][word] = 1
```

```python
word_freq_processed
```

```python
top10 = {1:[], 2:[], 3:[], 4:[], 5:[]}
for i in range(1,6):
    word_freq_processed[i]['0'] = 0
    for j in range(10):
        top = max(word_freq_processed[i], key=word_freq_processed[i].get)
        top10[i].append((top,word_freq_processed[i][top]))
        word_freq_processed[i][top] = 0
```
